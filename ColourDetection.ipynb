```python
#import library OpenCV
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

#ekstraksi citra RED
img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/rgb.png',1)
print('Input image')
cv2_imshow(img)
Blue_Channel= img[:,:,2]
Green_Channel=img[:,:,1]
Red_Channel=img[:,:,0]
print('Red_Channel')
cv2_imshow(Red_Channel)


gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print('Gray Image')
cv2_imshow(gray)

Blue=cv2.subtract(Blue_Channel,gray)
print('Subtracted Image')
cv2_imshow(Blue)

ret,thresh=cv2.threshold(Blue,110,255,cv2.THRESH_BINARY)
print('Threshold Image')
cv2_imshow(thresh)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/gulfkanawut_94565919_2667730966794152_8761659056641007557_n.jpg',1)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

Blue_Low = np.array([110,50,50])
Blue_High = np.array([130,225,225])

mask = cv2.inRange(hsv, Blue_Low, Blue_High)

cv2_imshow(img)

cv2_imshow(mask)
```
